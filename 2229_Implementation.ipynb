---

## 시간 초과

````python
N, K = map(int, input().split())

lst = list(map(int, input().split()))

ans = 0

s = 0

for si in range(s, N - K + 1):
    ans = max(ans, sum(lst[si:si+K]))

print(ans)
````

---

## 시간 초과 해결

- 바로 뒷자리를 추가, 제일 앞을 제거해서 연산을 최소화

- max 처리시, ans = 0 으로 하면 error, 합이 음수만 있는 경우가 있으므로, 초기값 설정 중요

```

0 1 2 3 4 5 6 7

3 2 4 9 0 3


```

````python

# 1) 입력

N, K = map(int, input().split())

lst = list(map(int, input().split()))


# 2) 처리

ans = sum(lst[:K]) # 0으로 처리하면 x, 음수가 답인 경우가 존재, ans = 0만 나옴, 

s = sum(lst[:K]) 

for i in range(K, N):
    
    s = s + lst[i] - lst[i - K]
    
    ans = max(ans, s)

print(ans)
````
