4초 등장, 1초 기달



```python
N, K = map(int, input().split())


ans = 0

start = 0

for _ in range(N):
    
    lst = list(map(int, input().split()))
    
    
    light = lst[0]
    red = lst[1]
    blue = lst[2]
    
    # 다음 신호등 까지 이동 시간 을 더한 현재시간 
    ans += light - start 
    
    # 현재 시간 기준, 초록불 이라면, 
    if ans % (red + blue) >= red:
        pass
    
    #  현재 시간 기준, 빨간불 이라면, 
    else:
        ans +=  red - ( ans % (red + blue) )
    
    # 신호등 초기화
    start = light 
    
ans += K - start

print(ans)
    
```

---

데이터 타입 정의 후 처리

- 현재 신호등에서의 대기시간 추가
- 신호등주기(r+g)로 나머지 구하면 => 진입시점
- 진입시점이 r 보다 작으면 r시간이 끝날때까지 대기하므로 r - 진입시간

```python
N, L = map(int, input().split())

# 시작지점과 목적지(L)을 포함해서 모든 신호등 위치 저장

arr = [[0,0,1]] + [ list(map(int, input().split())) for _ in range(N) ] + [[L,0,1]] 

cnt = 0

for i in range(1, len(arr)):
    
    d, r, g = arr[i]
    
    # 지난 위치부터 현재위치의 차이(이동거리) 추가
    
    cnt += (d - arr[i-1][0])

    cnt += max(0, r-(cnt%(r+g)))
     
print(cnt)
```
