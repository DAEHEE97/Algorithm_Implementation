# [사탕게임_구현](https://www.acmicpc.net/problem/3085)

- 구현

- 브루트포스 알고리즘

- cnt_consecutive

- 가능한 모든 경우 처리 후 , 후 최대 길이 갱신

- 처리 방향, 오른쪽, 아래쪽 (Transpose)

- 카운트 처리 : 단순 카운트가 아니라, 몇개 가 붙어 있는지 검사 해야 함


---

```python
# 행 개수 카운트, 붙어 있는

def count(lst):
    
    cnt = 1
    ans = 1
    
    for i in range(1, len(lst)):
        
        if lst[i] == lst[i-1]:
            cnt += 1
            ans = max(ans, cnt)
        
        else:
            cnt = 1
    
    return ans


def solve(arr):
    
    mx = 0
    
    for i in range(0, N-1):
        
        for j in range(0, N):
            
            # 오른쪽 사탕과 교환
            
            arr[i][j], arr[i][j+1] = arr[i][j+1], arr[i][j]
            
            mx = max(mx, count(arr[i]))
            
            arr[i][j], arr[i][j+1] = arr[i][j+1], arr[i][j] # 원래대로 값 변경  

            # 아래쪽 사탕과 교환
            
            arr[i][j], arr[i+1][j] = arr[i+1][j], arr[i][j]
            
            mx = max( mx, count(arr[i]), count(arr[i+1]) ) # 위 아래 변경, 2줄 모두 검사 
            
            arr[i][j], arr[i+1][j] = arr[i+1][j], arr[i][j] # 원래대로 값 변경  
    
    return mx


N = int(input())

arr = [ list(input()) + [0] for _ in range(N) ] + [ [0] * (N+1) ]

arr_t = [ list(tup) for tup in zip(*arr) ]

# arr_t = list(map(list, zip(*arr)))

ans = max(solve(arr), solve(arr_t))

print(ans)
```

```python
# consecutive cnt

def find_max_consecutive(lst):
    
    max_count = 1  # 최대 개수를 저장할 변수
    
    cnt = 1  # 현재 연속 개수를 저장할 변수

    for i in range(1, len(lst)):
        
        if lst[i] == lst[i-1]:
            
            cnt += 1
            
            if cnt > max_count:
                max_count = cnt
        
        else:
            cnt = 1 # 전 값과 다르다면, cnt = 1 처리, 이전에 이미 max 를 이미 저장 한 상태

    return max_count
```
